```python
df["total_scre"]=df.language+df.science+df.maths+df.history
```

```python
df
```

```python
df["score %"]=df.total_scre/4
```

```python
def gread(g_d):
    if(g_d>60):
        return "A"
    elif(g_d<60 and g_d>=45):
        return "B"
    else:
        return "C"
```

```python
df["gread"]=df["score %"].apply(gread)
```
